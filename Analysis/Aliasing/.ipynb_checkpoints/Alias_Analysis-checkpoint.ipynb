```python
import numpy as np
import lightkurve as lk
from matplotlib import pyplot as plt
from copy import deepcopy
```

# Window Function for Zvrk

```python
search_result = lk.search_lightcurve('TIC 350842552', author='SPOC', exptime=120)
search_result
```

```python
lc_collection = search_result.download_all()
lc_stitched = lc_collection.stitch()
lc_stitched = lc_stitched[lc_stitched.quality == 0]
```

```python
import astropy.units as u
from copy import deepcopy

lc = deepcopy(lc_stitched)
# lc = lc[ (lc.time.value > 3050) * (lc.time.value > 3070)]
dur = lc.time.value[-1] - lc.time.value[0] # days
dt = np.median(np.diff(lc.time.value)) * u.d
dur_ts = lc.time.value - lc.time.value[0]

time_array = np.arange(0, dur, dt.value) 
win_array = np.ones_like(time_array)
time_array, dt



gap = np.where(np.diff(lc.time.value) > 0.5)[0] ## Find gaps larger than half a day


for idx in gap: # Finding gap indices
  idx_low = np.argmin(np.abs(time_array - dur_ts[idx] ))
  idx_hi = np.argmin(np.abs(time_array - dur_ts[idx + 1] ))
  win_array[idx_low:idx_hi] = 0



fft_result = np.fft.fft(win_array)
spectral_window = np.abs(fft_result)


fig = plt.figure(figsize=(12,6))
ax1, ax2 = fig.add_subplot(121), fig.add_subplot(122)

ax1.plot(time_array, win_array)
ax1.set_title('Observational Window')

# Length of the time series
N = len(time_array)

# sample rate should be the number of samples in a day. 1 day has 1440 minutes, so there are 720 samples per day at a 2-min cadence.
fs = 720 

# Create the frequency axis
frequencies = np.fft.fftfreq(N, 1/fs) # units of cycles per day
cd_to_Hz = (1/(1*u.d)).to(u.uHz).value

ax2.plot(frequencies*cd_to_Hz, spectral_window)
ax2.set_xlabel('Frequency (uHz)')
ax2.set_ylabel('Spectral Window Amplitude')
ax2.set_title('Spectral Window Function')
ax2.set_xlim(-2, 2)
ax2.axvline(x=0.85, c='r', ls='--', alpha=0.25, label= 'Approx.\n1 day downlink')
ax2.legend(loc='upper left')
plt.show()
```

## Looking at Fafnir's PSD -> next part needs echelle

```python
from echelle import plot_echelle

ax = plot_echelle(freq = pg_ori.frequency.value, power= pg_ori.power.value, dnu =1.56, fmin=5, fmax=15, offset = 0,
               cmap='jet')
plt.show()
```
